```python
import pandas as pd
import numpy as np
db = pd.read_excel('Consolidado.xlsx')
```

```python
db.head(5)
```

```python
lista = []
for x in db.columns:
    if db[x].dtype == 'object':
        lista.append(x)
lista
```

```python
lista.remove('Paises')
```

```python
lista
```

```python
db = db.replace({'--':np.NaN})
```

```python
for x in db.columns:
    if x in lista:
        db[x].astype(float)
```

```python
db.info()
```

```python
lista_agg = []
for x in db.columns:
    if db[x].dtype == float:
        lista_agg.append(x)
lista
```
